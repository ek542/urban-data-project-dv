```python
# Populate empty DFs with sum of RR + RC + FAC
def populate_totals(total_year, dv_year):
    dv_year = dv_year.iloc[1:, :]
    total = total_year
    total['precinct'] = dv_year['Jan ']
    total['jan'] = dv_year['Unnamed: 1'] + dv_year['Unnamed: 2'] + dv_year['Unnamed: 3']
    total['feb'] = dv_year['Unnamed: 5'] + dv_year['Unnamed: 6'] + dv_year['Unnamed: 7']
    total['march'] = dv_year['Unnamed: 9'] + dv_year['Unnamed: 10'] + dv_year['Unnamed: 11']    
    total['april'] = dv_year['Unnamed: 13'] + dv_year['Unnamed: 14'] + dv_year['Unnamed: 15']
    total['may'] = dv_year['Unnamed: 17'] + dv_year['Unnamed: 18'] + dv_year['Unnamed: 19']
    total['june'] = dv_year['Unnamed: 21'] + dv_year['Unnamed: 22'] + dv_year['Unnamed: 23']
    total['july'] = dv_year['Unnamed: 25'] + dv_year['Unnamed: 26'] + dv_year['Unnamed: 27']
    total['aug'] = dv_year['Unnamed: 29'] + dv_year['Unnamed: 30'] + dv_year['Unnamed: 31']
    total['sept'] = dv_year['Unnamed: 33'] + dv_year['Unnamed: 34'] + dv_year['Unnamed: 35']    
    total['oct'] = dv_year['Unnamed: 37'] + dv_year['Unnamed: 38'] + dv_year['Unnamed: 39']
    total['nov'] = dv_year['Unnamed: 41'] + dv_year['Unnamed: 42'] + dv_year['Unnamed: 43']
    total['dec'] = dv_year['Unnamed: 45'] + dv_year['Unnamed: 46'] + dv_year['Unnamed: 47']
    return total
```

```python
# Populate empty DFs with sum of RR + RC + FAC
df_2017 = populate_totals(total_2017, dv_2017)
df_2018 = populate_totals(total_2018, dv_2018)
df_2019 = populate_totals(total_2019, dv_2019)
df_2020 = populate_totals(total_2020, dv_2020)
```

### Make Prophet-ready

```python
# Function to populate DF with ds in datetime format 
import datetime
def create_dates(year):
    lst = []
    for i in range(1, 13):
        lst.append(pd.to_datetime(datetime.datetime(year, i, 1)))
    return lst
```

```python
# Create Prophet ready DFs

p_2017 = pd.DataFrame(columns=['ds', 'y'])
p_2017['ds'] = create_dates(2017)
p_2017['y'] = cases_2017

p_2018 = pd.DataFrame(columns=['ds', 'y'])
p_2018['ds'] = create_dates(2018)
p_2018['y'] = cases_2018

p_2019 = pd.DataFrame(columns=['ds', 'y'])
p_2019['ds'] = create_dates(2019)
p_2019['y'] = cases_2019

p_2020 = pd.DataFrame(columns=['ds', 'y'])
p_2020['ds'] = create_dates(2020)
p_2020['y'] = cases_2020
```

### Prophet Validation

```python
# Test Prophet forecast for 2019 using 2017, 2018 data
p_train_2017_2018 = pd.concat([p_2017, p_2018])

m = Prophet(yearly_seasonality=True)
m.add_country_holidays(country_name='US')
m.fit(p_train_2017_2018)

p_test_2019 = pd.DataFrame(columns=['ds'], data=create_dates(2019))

forecast_2019 = m.predict(p_test_2019)
```

### Per Precinct Models

```python
df_2017_T = df_2017.T.iloc[:,0:77]
df_2017_T.columns = df_2017_T.iloc[0]
df_2017_T.drop(df_2017_T.index[0], inplace=True)
df_2017_T['ds'] = create_dates(2017)
df_2017_T.set_index('ds', inplace=True)

df_2018_T = df_2018.T.iloc[:,0:77]
df_2018_T.columns = df_2018_T.iloc[0]
df_2018_T.drop(df_2018_T.index[0], inplace=True)
df_2018_T['ds'] = create_dates(2018)
df_2018_T.set_index('ds', inplace=True)

df_2019_T = df_2019.T.iloc[:,0:77]
df_2019_T.columns = df_2019_T.iloc[0]
df_2019_T.drop(df_2019_T.index[0], inplace=True)
df_2019_T['ds'] = create_dates(2019)
df_2019_T.set_index('ds', inplace=True)

df_2020_T = df_2020.T.iloc[:,0:77]
df_2020_T.columns = df_2020_T.iloc[0]
df_2020_T.drop(df_2020_T.index[0], inplace=True)
df_2020_T['ds'] = create_dates(2020)
df_2020_T.set_index('ds', inplace=True)
```

```python
df_2020_T.columns[-1]
```

```python
forecast_precincts = pd.DataFrame(columns=['dates'], data=create_dates(2020)[2:])
def prophet_precinct(df_2017_T, df_2018_T, df_2019_T, df_2020_T):
    for i in range(77):
        train_2017 = df_2017_T.iloc[:, i]
        train_2018 = df_2018_T.iloc[:, i]
        train_2019 = df_2019_T.iloc[:, i]
        train_2020 = df_2020_T.iloc[:2, i]
        train = pd.concat([train_2017, train_2018, train_2019, train_2020]).to_frame()
        
        col = train.columns[-1]
        train.rename(columns={col: 'y'}, inplace=True)
        train.reset_index(inplace=True)
        #print(train)
        
        test = pd.DataFrame(columns=['ds'], data=create_dates(2020)[2:])
        
        m = Prophet(yearly_seasonality=True)
        m.fit(train)

        forecast = m.predict(test)
        
        forecast_precincts[str(i)] = forecast['yhat']
    
    return forecast_precincts
```
